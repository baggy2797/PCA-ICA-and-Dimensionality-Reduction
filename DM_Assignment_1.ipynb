```
import pandas as pd 
data1 = pd.read_csv("dist1_500_1.txt",sep=" ", header=None) 

data2 = pd.read_csv("dist1_500_2.txt",sep=" ", header=None) 


dt1 = pd.DataFrame(data1)
dt2 = pd.DataFrame(data2)

#print(dt1)
dt_1 = dt1.dropna(how='all')
dt_2 = dt2.dropna(how='all')
#print(dt_1)
#print(dt_2)

data3 = pd.read_csv("dist2_500_1.txt",sep=" ", header=None) 
data4 = pd.read_csv("dist2_500_2.txt",sep=" ", header=None) 

dt3 = pd.DataFrame(data3)
dt4 = pd.DataFrame(data4)

dt_3 = dt3.dropna(how='all')
dt_4 = dt4.dropna(how='all')
```

```
#print(dt_2.tail(10))
df_row_1 = pd.concat([dt_1, dt_2])
df_row_2 = pd.concat([dt_3, dt_4])



df_row_1.to_csv('out1.csv',index=False)
df_row_2.to_csv('out2.csv',index=False)


#taking ten random points from the file
DATA1 = df_row_1.sample(n=10)


DATA2 = df_row_2.sample(n=10)

#################################
import matplotlib.pyplot as plt
import pandas as pd

from statsmodels.graphics.gofplots import qqplot
from statsmodels.graphics.gofplots import qqplot
from matplotlib import pyplot

for i in range(10) :
  plt.hist(DATA1.iloc[i])
plt.show()

for i in range(10) :
  plt.hist(DATA2.iloc[i])
plt.show()


print('DATA1')
qqplot(DATA1, line='s')
plt.show()

print('DATA2')
qqplot(DATA2, line='s')
plt.show()


qqplot(DATA1.iloc[5],line='s')
qqplot(DATA2.iloc[4],line='s')
plt.show()
```

```
print("-------------Mean :1--------------")

print(DATA1.mean(axis=1))

print("-------------Mean :2--------------")
print(DATA2.mean(axis=1))

print("-------------Mode :1--------------")

print(DATA1.mode(axis=1))

print("-------------Mode :2--------------")

print(DATA2.mode(axis=1))

print("-------------Median :1-------------") 

print(DATA1.median(axis=1))

print("-------------Median :2-------------") 

print(DATA2.median(axis=1))


print("-------------Variance :1------------")

print(DATA1.var(axis = 1))

print("-------------Variance :2------------")

print(DATA2.var(axis = 1))

print("-------------Standard Deviation :1----------")

print(DATA1.std(axis=1))

print("-------------Standard Deviation :2----------")

print(DATA2.std(axis=1))
```

```
import numpy as np
from sklearn.decomposition import PCA

x_std = df_row_1
for i in range(100):
  pca = PCA(n_components=i)
  sklearn_pca_x = pca.fit_transform(x_std)
  sklearn_result = pd.DataFrame(sklearn_pca_x)
  sklearn_result['y-axis'] = 0.0
  sklearn_result['label'] = 0
  if sum(pca.explained_variance_ratio_) > 0.90:
    print(i)
    break

plt.scatter(sklearn_pca_x[:,0],sklearn_pca_x[:,1])
plt.show()

from sklearn.decomposition import FastICA
transformer = FastICA(n_components=100,random_state=0)
data_transformed = transformer.fit_transform(df_row_1)

#plt.hist(data_transformed)
#plt.show()
plt.scatter(data_transformed[:,0],data_transformed[:,1])
plt.show()
```
